```python
## imports
import numpy as np
import matplotlib.pyplot as plt

np.warnings.filterwarnings("ignore")
```

```python
def mandel(n, c):
    z = 0
    for _ in range(n):
        z = z ** 2 + c
    return np.abs(z) <= 2
```

```python
dpi = 512
ax_re = np.linspace(-2, 1, dpi)
ax_im = np.linspace(-1.5, 1.5, dpi)
grid = ax_re[np.newaxis, :] + ax_im[:, np.newaxis] * 1j

plt.imshow(mandel(20, grid), cmap="binary")
plt.axis('off')
plt.show()
```
